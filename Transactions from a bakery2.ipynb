```python

secondhalf_of_2016 = data.loc[(data['Date']>='2016-06-01') &(data['Date']<='2016-12-31')]
firsthalf_of_2017 = data.loc[(data['Date']>='2017-01-01') &(data['Date']<='2017-05-31')]


year = {'second2016' :len(secondhalf_of_2016), 'first2017' : len(firsthalf_of_2017) }
print("this is the first half of 2017 : ", year['first2017'])
```

```python

plt.figure(figsize=(5,5))
plt.ylabel('Values', fontsize='medium')
plt.xlabel('Items', fontsize='medium')
plt.title('Top 20 Sell Bakery Items')
plt.bar(list(year.keys()), list(year.values()))
plt.show()
```
